```python
import pandas as pd
import glob
import re
```

```python
def remove_emoji(string):
    emoji_pattern = re.compile("["
                           u"\U0001F600-\U0001F64F" # emoticons
                           u"\U0001F300-\U0001F5FF" # symbols & pictographs
                           u"\U0001F680-\U0001F6FF" # transport & map symbols
                           u"\U0001F1E0-\U0001F1FF" # flags (iOS)
                           u"\U00002702-\U000027B0" # Dingbats
                           u"\U0001F700-\U0001FBFF" # supplement symbols
                           "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', string)
```

## Processing

```python
data = pd.read_csv('data.csv')
data
```

```python
data.drop_duplicates(inplace = True)
data.dropna(subset = ['review'], inplace = True)
```

```python
with open('./Notes/spam.txt','r', encoding='utf-8') as file:
    spam = file.readlines()
```

```python
spam = [remove_emoji(s.strip('\n')) for s in spam]
```

```python
len(spam)
```

```python
while 1:
    try:
        spam.remove('')
    except:
        print(len(spam))
        break
```

```python
data.shape
```

```python
for row in data.itertuples():
    if row.review in spam:
        data.drop(index = row.Index, inplace = True)
```

```python
data.shape
```

```python
data.groupby('rating').count()
```

```python
pattern = re.compile(r'(?<!\d)\.+|\.+(?!\d)|\n')
```

```python
data['review'] = data.review.apply(lambda x: pattern.split(x))
```

```python
data = data.explode('review', ignore_index= True)
```

```python
data
```

```python
data['len'] = data.review.apply(lambda x: len(x.split()))
```

```python
data
```

```python
data = data[data.len>5].copy()
```

```python
data.drop(columns='len', inplace = True)
```

```python
data.shape
```

```python
print('Before remove shop related sentences')
data.groupby('rating').count()
```

```python
with open('./Notes/shop.txt','r', encoding='utf-8') as file:
    drop_shop = file.readlines()
```

```python
drop_shop = [s.strip('\n') for s in drop_shop]
```

```python
len(drop_shop)
```

```python
for row in data.itertuples():
    for item in drop_shop:
        if item in row.review:
            data.drop(index = row.Index, inplace = True)
            break
```

```python
data.shape
```

```python
print('After remove shop related sentences')
data.groupby('rating').count()
```

```python
with open('./Notes/abbreviation.txt','r', encoding='utf-8') as file:
    abbr = file.readlines()
with open('./Notes/fullWord.txt','r', encoding='utf-8') as file:
    full_word = file.readlines()
```

```python
for i, j in zip(abbr,full_word):
    i = i.strip('\n').lower()
    j = j.strip('\n').lower()
    # print(i)
    # print(j)
    for k in data.index:
        pattern2 = re.compile(f'(?<!\w){i}(?!\w)')
        s = data.loc[k,'review']
        s = pattern2.sub(j,s.lower())
        data.loc[k,'review'] = s
```
