```python
def ROTR(e, t):
    return t >> e | t << 32 - e

def Sigma0(e):
    return ROTR(2, e) ^ ROTR(13, e) ^ ROTR(22, e)

def Sigma1(e):
    return ROTR(6, e) ^ ROTR(11, e) ^ ROTR(25, e)

def sigma0(e):
    return ROTR(7, e) ^ ROTR(18, e) ^ e >> 3

def sigma1(e):
    return ROTR(17, e) ^ ROTR(19, e) ^ e >> 10

def Ch(e, t, r):
    return e & t ^ ~e & r

def Maj(e, t, r):
    return e & t ^ e & r ^ t & r

def toHexStr(e):
    t = ''
    for r in [7, 6, 5, 4, 3, 2, 1, 0]:
        temp = (e >> 4 * r & 15)
        t += '{0:x}'.format(temp)
    return t

def from_char_code(*args):
    return ''.join(map(chr, args))

def charCodeAt(string, index):
    if index >= len(string):
        return 0
    if isinstance(string, str) and isinstance(index, int):
        return ord(string[index])
    else:
        raise Exception(
            f"Error: charcodeat takes a str and an int, received {type(string)} and {type(int)}.")

def hash(e, n=0):
    a = [1116352408, 1899447441, 3049323471, 3921009573, 961987163, 1508970993, 2453635748, 2870763221, 3624381080, 310598401, 607225278, 
        1426881987, 1925078388, 2162078206, 2614888103, 3248222580, 3835390401, 4022224774, 264347078, 604807628, 770255983, 1249150122,
        1555081692, 1996064986, 2554220882, 2821834349, 2952996808, 3210313671, 3336571891, 3584528711, 113926993, 338241895, 666307205,
        773529912, 1294757372, 1396182291, 1695183700, 1986661051, 2177026350, 2456956037, 2730485921, 2820302411, 3259730800, 3345764771, 
        3516065817, 3600352804, 4094571909, 275423344, 430227734, 506948616, 659060556, 883997877, 958139571, 1322822218, 1537002063, 
        1747873779, 1955562222, 2024104815, 2227730452, 2361852424, 2428436474, 2756734187, 3204031479, 3329325298]
    o = [1779033703, 3144134277, 1013904242, 2773480762, 1359893119, 2600822924, 528734635, 1541459225]
    e += from_char_code(128)
    l = len(e) / 4 + 2
    i = math.ceil(l / 16)
    c = [0]*16

    for u in range(i):
        c[u] = [0]*16
        for s in range(0, 16):
            c[u][s] = charCodeAt(e, 64 * u + 4 * s) << 24 | \
                charCodeAt(e, 64 * u + 4 * s + 1) << 16 | \
                charCodeAt(e, 64 * u + 4 * s + 2) << 8 | \
                charCodeAt(e, 64 * u + 4 * s + 3)
    c[i - 1][14] = 8 * (len(e) - 1) / math.pow(2, 32)
    c[i - 1][14] = math.floor(c[i - 1][14])
    c[i - 1][15] = 8 * (len(e) - 1) & 4294967295
    # m = p = d = f = h = y = g = v = b = [[]]*64
    b = [[]]*64
    for u in range(i):
        for E in range(0, 16):
            b[E] = c[u][E]
        for E in range(16, 64):
            b[E] = sigma1(b[E - 2]) + b[E - 7] + \
                sigma0(b[E - 15]) + b[E - 16] & 4294967295
        m = o[0]
        p = o[1]
        d = o[2]
        f = o[3]
        h = o[4]
        y = o[5]
        g = o[6]
        v = o[7]
        for E in range(0, 64):
            N = v + Sigma1(h) + Ch(h, y, g) + a[E] + b[E]
            S = Sigma0(m) + Maj(m, p, d)
            v = g
            g = y
            y = h
            h = f + N & 4294967295
            f = d
            d = p
            p = m
            m = N + S & 4294967295

        o[0] = o[0] + m & 4294967295
        o[1] = o[1] + p & 4294967295
        o[2] = o[2] + d & 4294967295
        o[3] = o[3] + f & 4294967295
        o[4] = o[4] + h & 4294967295
        o[5] = o[5] + y & 4294967295
        o[6] = o[6] + g & 4294967295
        o[7] = o[7] + v & 4294967295
    return toHexStr(o[0]) + toHexStr(o[1]) + toHexStr(o[2]) + toHexStr(o[3]) + toHexStr(o[4]) + toHexStr(o[5]) + toHexStr(o[6]) + toHexStr(o[7])
```

```python
def get_s_suffix(params):
    if params.get('s'):
        params.pop('s')
    params_ls = list(params.items())
    params_str_ls =[]
    for i in params_ls:
        params_str_ls.append('='.join(i))
    params_str = '&'.join(params_str_ls)
    return params_str
```

```python
def get_cmt():
    cmt_ls = []
    old_cmt = ''
    while True:
        sleep(random.randint(2,6))
        try:
            cmts_container = driver.find_element(By.CSS_SELECTOR, '.shopee-product-comment-list')
        except:
            break
        
        if cmts_container.text == old_cmt:
            break
        else:
            old_cmt = cmts_container.text

        cmts = cmts_container.find_elements(By.CSS_SELECTOR, '._280jKz')
        for cmt in cmts:
            try:
                cmt_ls.append(cmt.find_elements(By.CSS_SELECTOR, 'div')[-1].text)
            except:
                cmt_ls.append(cmt.text)

        next_page = driver.find_element(By.CSS_SELECTOR, '.shopee-icon-button.shopee-icon-button--right').click()
    return cmt_ls
```
